```python
import matplotlib as plt
import numpy as np
import phaseportrait
```

```python
def pendulum(, d):
    return d, - (np.sin())**2

Pendulum = phaseportrait.PhasePortrait2D(pendulum, [-9, 9], Title='Pendulum', xlabel=r"$\Theta$", ylabel=r"$\dot{\Theta}$")
Pendulum.plot()
```

```python
def oscillator(x,dx):
    return dx, -2*x**2+x-4

Oscillator = phaseportrait.PhasePortrait2D(oscillator, [-100,100], Title = 'Oscillator')
Oscillator.plot()
```

```python
def nonlinear(x,dx):
    return dx,10*(x**2)*(1-x);  
    
Nonlinear = phaseportrait.PhasePortrait2D(nonlinear, [-3,3], Title = "Oscillator-Nonlinear")
Nonlinear.plot()
```

```python
def function(y,dx):
    return dx, -x*y+x**3*(1-y**2)

Function = phaseportrait.PhasePortrait2D(function, [-10,10], Title = "Coupled equation")
Function.plot()
```
